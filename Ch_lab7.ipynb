```python
import numpy as np
import matplotlib.pyplot as plt

x0 = 0.0      #начальная точка
y0 = 1.0      #начальное значение
x_end = 2.0   #конечная точка
h = 0.1       #шаг сетки

def f(x, y):
    return x**2

def y_analytic(x):
    return (x**3)/3 + 1

def euler_method(f, x0, y0, x_end, h):
    #создаю сетку
    x_values = np.arange(x0, x_end + h, h)
    y_values = np.zeros(len(x_values))
    y_values[0] = y0
    
    #применяю метод Эйлера
    for i in range(1, len(x_values)):
        y_values[i] = y_values[i-1] + h * f(x_values[i-1], y_values[i-1])
    
    return x_values, y_values

#численное решение
x_num, y_num = euler_method(f, x0, y0, x_end, h)
y_anal = y_analytic(x_num)

print(" x   | Численное y | Аналитическое y | Погрешность")
for x, yn, ya in zip(x_num, y_num, y_anal):
    print(f"{x:.1f}  | {yn:.6f}    | {ya:.6f}        | {abs(yn-ya):.6f}")

plt.figure(figsize=(10, 6))
plt.plot(x_num, y_num, 'b-', label='Численное решение (Эйлер)')
plt.plot(x_num, y_anal, 'r--', label='Аналитическое решение')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Сравнение численного и аналитического решений')
plt.legend()
plt.grid(True)
plt.show()
```
