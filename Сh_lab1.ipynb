```python
#проверка компьютера на погрешность вычисления
def test_exp():
    k = 0  
    for n in range(1000):  
        eps = 2**(-n)  
        res = (1+eps-1)/eps 
        k+=1 
        print(f'ШАГ {k}: eps = {eps}, результат = {res}')
        if res!=1:  
            print(f'Результат перестал быть равным 1 при n = {n}')
            break

test_exp()
#в мантиссу помещается не более 15-17 знаков, из-за этого происходит округление и результат не равен 1
```

```python
I = [math.exp(-1)]  
for n in range(2, 31):
    I_n = 1 - n * I[-1]  
    I.append(I_n)
    
for n, I_n in enumerate(I, start=1):
    print(f'I_{n} = {I_n}')

#вывод: данный феномен происходит из-за накопления погрешностей
```

# Задание 3

```python
eps = 1e-17
points = [0, np.pi/3, np.pi/2, np.pi, 2*np.pi, 12*np.pi, 13*np.pi, 14*np.pi]

for x in points:
    summ = 0
    u_n = x
    n = 0
    while abs(u_n) >= eps:
        summ += u_n
        n += 1
        u_n *= (-x**2)/((2*n) * (2*n + 1))
        if x in [12*np.pi, 13*np.pi, 14*np.pi]:
            print(f'x = {x}, шаг {n}: член ряда = {u_n}, частичная сумма = {summ}')
    print(f'x = {x}, вычисленное значение sin(x) = {summ}, точное значение sin(x) = {np.sin(x)}')
    print('')
```
